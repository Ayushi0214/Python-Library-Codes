```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()

driver.get("""https://facilities.aicte-india.org/dashboard/pages/angulardashboard.php#!/approved""")

time.sleep(2)

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[1]/select""").click()

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[1]/select/option[2]""").click()

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[2]/select""").click()

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[2]/select/option[10]""").click()

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[3]/select""").click()

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[3]/select/option[1]""").click()

course = driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/div[5]/input""")
course.send_keys("food technology")

time.sleep(1)

driver.fullscreen_window()

time.sleep(1)

driver.find_element(By.XPATH, """/html/body/div[2]/div/nav/div[2]/button""").click()

soup = BeautifulSoup(driver.page_source, "html.parser")
soup
```
