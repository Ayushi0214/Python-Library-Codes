```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")
df
```

```python
x = [1,2,3,4,5,6,7]
y = [90,80,75,80,77,67,90]

plt.figure(figsize = (4,3))
plt.bar(x, y, color = "pink", edgecolor = "black", linestyle = ":")
plt.xlabel("Parts of Harry Potter",color = "g")
plt.ylabel("Popularity", color = 'g')
plt.title("Popularity of all the parts of Harry Potter")
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
gb = df.groupby("day").agg({"total_bill":"sum"})
plt.bar(gb.index, gb["total_bill"])
plt.show()
```
