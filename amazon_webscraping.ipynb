```python
import requests
from bs4 import BeautifulSoup
Names = []
Ratings = []
Prices = []
Boughts = []
OPs = []

url = "https://www.amazon.in/s?k=phone+under+50000&crid=2NI5BROZ2X1O4&sprefix=phone+under+50000%2Caps%2C278&ref=nb_sb_noss_2"

for i in range(2,11):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    
    name = soup.find_all("span", {"class":"a-size-medium a-color-base a-text-normal"})
    
    for i in name:
        Names.append(i.text)
    
    price = soup.find_all("span", {"class":"a-price-whole"})
    
    for i in price:
        Prices.append(i.text)
    
    rating = soup.find_all("span", {"class":"a-icon-alt"})
    
    for i in rating:
        Ratings. append(i.text)
    
    Ratings.pop(-1)
        
    link = soup.find("a", {"class":"s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"}).get("href")
    link = link[:-1]
    cnl = "https://www.amazon.in/s?k="+link+str(i)
    url = cnl
    
```

```python
import pandas as pd 

df = pd.DataFrame({"Names":Names, "Prices":Prices, "Ratings":Ratings})
df
```

```python
df.to_csv("amazon_phones_under_50k.csv", index = False)
```
