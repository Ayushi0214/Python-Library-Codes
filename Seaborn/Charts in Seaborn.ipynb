```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.barplot(x = "day", y = "tip", data = df, estimator = "sum", errorbar = ("ci", 95), hue = "smoker", palette = "viridis",)
plt.title("avg tips by days in a week")
plt.show()
```

```python
df = sns.load_dataset("taxis")
df
```

```python
df["hour"] = df["pickup"].dt.strftime("%H")
df = df.sort_values(by = "hour")
df
```

```python
sns.lineplot(x = "hour", y = "total", data = df, hue = "payment" )
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the sample dataset
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F', 'Product G'],
    'Rating': [4.5, 3.8, 4.2, 3.9, 4.0, 4.7, 3.5],
    'Likes': [120, 95, 80, 150, 60, 110, 130],
    'Sales (Units)': [500, 400, 600, 350, 450, 550, 200],
    'Price ($)': [49.99, 19.99, 59.99, 29.99, 24.99, 39.99, 69.99]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df = df[["Rating","Likes","Sales (Units)","Price ($)" ]]
df
```

```python
sns.heatmap(data = df.corr(), annot = True)
plt.show()
```
