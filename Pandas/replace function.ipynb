```python
import pandas as pd

data = {"A":["a","b","c","d","e"],
       "job title":["senior manger", "analyst","junior manger","manger","analytics manger"],
       "gender":["male","fmale","male","fmale","fmale"]}

df = pd.DataFrame(data)
df
```

```python
df["job title"] = df["job title"].replace("manger","manager")
df
```

```python
df["job title"] = df['job title'].str.replace("manger","manager")
df
```

```python
df["gender"] = df["gender"].str.replace("fmale","female")
df
```
