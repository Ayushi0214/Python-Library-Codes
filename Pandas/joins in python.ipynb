```python
import pandas as pd
import numpy as np

# Sample data for products
products_data = {
    'Product ID': [101, 102, 103, 104, 105, 106],
    'Product Name': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E', 'Product F'],
    'Price': [15.00, 20.00, 12.50, 18.00, 25.00, 30.00],
    'Category': ['Electronics', 'Home Goods', 'Electronics', 'Clothing', 'Home Goods', 'Clothing']
}

# Create a DataFrame
products= pd.DataFrame(products_data)

# Display the DataFrame
products
```

```python
sales = pd.DataFrame({
    'Sale ID': [1, 2, 3, 4, 5, 6, 7],
    'Product ID': [101, 102, 101, 104, 105,102, 104],
    'Quantity Sold': [5, 2, 3, 8, 1, 6, 2],
    'Sale Date': ['2024-01-01', '2024-01-05', 
                  '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10']
})
sales
```

```python
df = pd.merge(right = sales, left = products, on = "Product ID", how = "inner")
df
```

```python
df = pd.merge(right = sales, left = products, on = "Product ID", how = "left")
df
```

```python
df = pd.merge(right = sales, left = products, on = "Product ID", how = "outer")
df
```

```python
df = pd.merge(right = sales, left = products, how = "cross")
df
```
