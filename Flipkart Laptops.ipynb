```python
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
import requests
import pandas as pd
import seaborn as sns

Names = []
Original_Price = []
Discount_Price = []
Discount = []
Ratings = []
No_of_Ratings = []
Processor = []
Storage = []
OS = []
RAM = []
Display = []
Warranty = []
url = "https://www.flipkart.com/search?q=laptop+under+70000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page="

for i in range(1,11):
    r = requests.get(url+str(i))
    soup = BeautifulSoup(r.text, "html.parser")
    main = soup.find("div",{"class":"DOjaWF gdgoEp"})

    sub = main.find_all("div",{"class":"cPHDOP col-12-12"})
    
    for i in sub:
        try:
            Name = i.find("div",{"class":"KzDlHZ"})
            Names.append(Name.text)
        except:
            Names.append(np.nan)
    
        try:
            p = i.find("div",{"class":"yRaY8j ZYYwLA"})
            Original_Price.append(p.text)
        except:
            Original_Price.append(np.nan)
    
        try:
            P = i.find("div",{"class":"Nx9bqj _4b5DiR"})
            Discount_Price.append(P.text)
        except:
            Discount_Price.append(np.nan)
    
        try:
            rating = i.find("div",{"class":"XQDdHH"})
            Ratings.append(rating.text)
        except:
            Ratings.append(np.nan)
    
        try:
            a = i.find("span",{'class':"Wphh3N"})
            No_of_Ratings.append(a.text)
        except:
            No_of_Ratings.append(np.nan)
    
        try:
            disc = i.find("div",{"class":"UkUFwK"})
            Discount.append(disc.text)
        except:
            Discount.append(np.nan)
    
        desp = i.find_all("li",{"class":"J+igdf"})
        try:
            Processor.append(desp[0].text)
        except:
            Processor.append(np.nan)
            
        try:
            RAM.append(desp[1].text)
        except:
            RAM.append(np.nan)
    
        try:
            OS.append(desp[2].text)
        except:
            OS.append(np.nan)
    
        try:
            Storage.append(desp[3].text)
        except:
            Storage.append(np.nan)
    
        try:
            Display.append(desp[4].text)
        except:
            Display.append(np.nan)
    
        try:
            abc = desp[5].text
            if "Warranty" not in abc:
                Warranty.append(desp[6].text)
            else:
                Warranty.append(abc)
            
        except:
            Warranty.append(np.nan)
    print(url+str(i))
```

```python
df = pd.DataFrame({"Names":Names, "Origial Price":Original_Price, "Discount_Price":Discount_Price, "Ratings":Ratings, "Processor":Processor, "Display":Display, "Warranty":Warranty, "RAM":RAM, "Storage":Storage, "OS":OS})
df
```
